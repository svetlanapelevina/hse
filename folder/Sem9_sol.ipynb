### Task 2

```python
url2 = 'https://raw.githubusercontent.com/justmarkham/DAT7/master/data/u.user'
```

```python
df = pd.read_table(url2, delimiter='|', header=None)
df.head()
# df.head(25)
```

```python
df.tail(10)
```

```python
df = df.iloc[:, 1:]
df.head()
```

```python
df.columns = ['age', 'gender', 'occupation', 'zip_code']
df.head()
```

```python
df.dtypes
```

```python
len(df.occupation.unique())
```

```python
df.occupation.value_counts()[:5]
```

```python
# df.groupby(['gender', 'occupation']).age.count()
```

```python
# df.describe()
# df.describe(include = 'all')
```

```python
df.age.mean()
```

```python
from scipy.stats import ttest_ind
ttest_ind(df.age[df.gender == 'M'], df.age[df.gender == 'F'])
```

```python
df.groupby('gender').age.mean()
```

```python
# d = df[df.occupation == 'student']
# (d.gender == 'M').sum() / d.gender.count()
```

```python
df.pivot_table(index = 'occupation', values='gender', 
    aggfunc=lambda x: (x == 'M').sum() / x.count())['gender'] \
    .sort_values(ascending=False)
```

```python
df.pivot_table(index = 'occupation', values='gender', 
    aggfunc=lambda x: (x == 'M').sum() / (x == 'F').sum())['gender'] \
    .sort_values(ascending=False)
```

```python
# df.groupby('occupation').age.mean()
```

```python
df.groupby(['occupation', 'gender']).agg(['mean', 'min', 'max'])
```

```python
from numpy import mean
from math import ceil
```

```python
df.pivot_table(index = ['gender', 'occupation'], values='age',  aggfunc=[lambda x: ceil(mean(x)), min, max])
```

```python
df.pivot_table(index='occupation', columns='gender', values='age', aggfunc=[mean, min, max])#.to_excel('sample.xlsx')
```

```python
a = df.groupby(['occupation', 'gender']).agg({'gender': 'count'})
b = df.groupby('occupation').agg({'gender': 'count'})

a.div(b, level='occupation')['gender']
```
